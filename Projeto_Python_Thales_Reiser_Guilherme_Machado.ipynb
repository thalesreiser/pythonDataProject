```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/content/owid-covid-data.csv', ',')
data
```

```python


pais1=data[data['location']=='Brazil']
pais2=data[data['location']=='China']

filtroDataBraz=pais1['date']
date=np.array(filtroDataBraz.head(30))


dadoCasePerMiBraz=pais1['total_cases_per_million']
dadoTotCaseBraz=pais1['total_cases']
dadoCasePerMiChin=pais2['total_cases_per_million']
dadoTotCaseChin=pais2['total_cases']
dadoTotDeathBraz=pais1['total_deaths']
dadoDeathPerMiBraz=pais1['total_deaths_per_million']
dadoTotDeathChin=pais2['total_deaths']
dadoDeathPerMiChin=pais2['total_deaths_per_million']

perMillionCasesBraz=np.array(dadoCasePerMiBraz.head(30))
perMillionDeathsBraz= np.array(dadoDeathPerMiBraz.head(30))
totalCasesBraz=np.array(dadoTotCaseBraz.head(30))
totalDeathBraz=np.array(dadoTotDeathBraz.head(30))
perMillionCasesChin=np.array(dadoCasePerMiChin.head(30))
totalCasesChin=np.array(dadoTotCaseChin.head(30))
totalDeathChin=np.array(dadoTotDeathChin.head(30))
perMillionDeathsChin=np.array(dadoDeathPerMiChin.head(30))

plt.figure(figsize=(30,14))

plt.plot(date, perMillionCasesChin, 'r.--',date, perMillionCasesBraz, 'b.:')
plt.title('Gráfico Brasil x China em número de casos por milhão no primeiro mês de pandemia')
plt.legend(['China', 'Brasil'])
plt.grid()


plt.show()
```

```python
plt.figure(figsize=(15,15))

plt.plot(date, totalCasesChin, 'r.--',date, totalCasesBraz, 'b.:')
plt.title('Gráfico Brasil x China em número de casos totais no primeiro mês de pandemia')
plt.legend(['China', 'Brasil'])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))

plt.plot(date, totalDeathChin, 'r.--',date, totalDeathBraz, 'b.:')
plt.title('Gráfico Brasil x China em número de mortes totais no primeiro mês de pandemia')
plt.legend(['China', 'Brasil'])
plt.grid()
plt.show()
```

```python
dataLast=np.array(filtroDataBraz.tail(30))

perMillionDeathsBrazLast= np.array(dadoDeathPerMiBraz.tail(30))
perMillionDeathsChinLast=np.array(dadoDeathPerMiChin.tail(30))

plt.figure(figsize=(15,15))

plt.plot(dataLast, perMillionDeathsChinLast, 'r.--',dataLast, perMillionDeathsBrazLast, 'b.:')
plt.legend(['China', 'Brasil'])
plt.title('Gráfico Brasil x China em número de mortes por milhão no último mês de pandemia')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,15))

plt.scatter(dataLast, perMillionCasesBraz, edgecolors='blue')
plt.scatter(dataLast, perMillionCasesChin, edgecolors='red')
plt.legend(['China', 'Brasil'])
plt.title('Gráfico Brasil x China em número de casos por milhão no último mês de pandemia')
plt
plt.grid()
plt.show()
```
